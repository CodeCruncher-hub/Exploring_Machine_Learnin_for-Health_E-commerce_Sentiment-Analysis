Naive Bayes using laplace correction

```python
from sklearn.naive_bayes import GaussianNB
import numpy as np

class LaplaceGaussianNB(GaussianNB):
    def fit(self, X, y, alpha=1.0):
        super().fit(X, y)
        self.class_prior_ = (self.class_count_ + alpha) / (self.class_count_.sum() + alpha * len(self.classes_))
        return self

# Use LaplaceGaussianNB instead of GaussianNB
clf = LaplaceGaussianNB()

# Train the classifier
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
import numpy as np
from sklearn import datasets #for inbuilt datasets
import torch #neural network implementation
import torch.nn as nn # for building and training neural networks
import torch.optim as optim #used to update the weights of neural network models
import torchbnn as bnn #library for Bayesian Neural Networks in PyTorch
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(-2, 2, 500) # 500 equally spaced values from -2 to 2 for generating synthetic data
y = x.pow(5) -10* x.pow(1) + 2*torch.rand(x.size()) #generates the output y based on a function of x random_noise = randomness to the data
#Different powers will result in different functions with different characteristics
x = torch.unsqueeze(x, dim=1) #reshapes the x tensor to have a shape of (500, 1).
y = torch.unsqueeze(y, dim=1) #reshapes the y tensor to have a shape of (500, 1).
#1D tensor of shape (500,) to a 2D tensor of shape (500, 1)
plt.scatter(x.data.numpy(), y.data.numpy())
plt.show()
```

```python
def clean_target(x):
    return x.pow(5) -10* x.pow(1)+1
def target(x):
    return x.pow(5) -10* x.pow(1) + 2*torch.rand(x.size())
```

```python
#nn.Sequential for layers
# two Bayesian neural network layers (bnn.BayesLinear) separated by a ReLU activation function (nn.ReLU)
model = nn.Sequential(
    bnn.BayesLinear(prior_mu=0, prior_sigma=0.1, in_features=1, out_features=1000),
    nn.ReLU(),
    bnn.BayesLinear(prior_mu=0, prior_sigma=0.1, in_features=1000, out_features=1),
)
#mean layer,SD,input feature,output
#This transformation allows the model to learn complex patterns and representations from the input data
# improving the model's capacity to learn from the data.
#prediction based on the complex representations learned in the preceding layers
```

```python
mse_loss = nn.MSELoss()
kl_loss = bnn.BKLLoss(reduction='mean', last_layer_only=False)
kl_weight = 0.01

optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
for step in range(2000):
    pre = model(x)
    mse = mse_loss(pre, y)
    kl = kl_loss(model)
    cost = mse + kl_weight*kl

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

print('- MSE : %2.2f, KL : %2.2f' % (mse.item(), kl.item()))
```

```python
x_test = torch.linspace(-2, 2, 300)
y_test = target(x_test)

x_test = torch.unsqueeze(x_test, dim=1)
y_test = torch.unsqueeze(y_test, dim=1)
```

```python
models_result = np.array([model(x_test).data.numpy() for k in range(10000)])
models_result = models_result[:,:,0]
models_result = models_result.T
mean_values = np.array([models_result[i].mean() for i in range(len(models_result))])
std_values = np.array([models_result[i].std() for i in range(len(models_result))])
```

```python
plt.figure(figsize=(10,8))
plt.plot(x_test.data.numpy(),mean_values,color='navy',lw=3,label='Predicted Mean Model')
plt.fill_between(x_test.data.numpy().T[0],mean_values-3.0*std_values,mean_values+3.0*std_values,alpha=0.2,color='navy',label='99.7% confidence interval')
#plt.plot(x_test.data.numpy(),mean_values,color='darkorange')
plt.plot(x_test.data.numpy(),y_test.data.numpy(),'.',color='darkorange',markersize=4,label='Test set')
plt.plot(x_test.data.numpy(),clean_target(x_test).data.numpy(),color='green',markersize=4,label='Target function')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
```
